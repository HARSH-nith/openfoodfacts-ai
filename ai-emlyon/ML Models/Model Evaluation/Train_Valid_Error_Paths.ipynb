# Train & Valid error paths with XGBoost

## Notebook config

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv(r'C:\Users\Antoine\Coding Bootcamp\Open Food Facts\Notebooks\Model Selection & Parameters\eval_results_xgboost.csv')
df.describe()
```

## Graphics

### XGBoost $G_1$ - Almost no regularization

```python
ann = r"$\min(L\log):$" 
ann += "\n"
ann += r"$train = 0.1714$"
ann += "\n"
ann += r"$valid = 0.4080$"
plt.figure(figsize=(12,6))
sns.lineplot(data=df[['G1_m0_train', 'G1_m0_val']], palette='viridis', linewidth=2.5)
plt.ylabel(r'$L_\log$ (Cross-Entropy)')
plt.xlabel(r'epoch ($N$ estimators)')
plt.title(r'XGBoost $G_1$ - Almost no regularization', fontsize=14)
plt.annotate(ann, xy =(435, 1.1), fontsize=11)
```
